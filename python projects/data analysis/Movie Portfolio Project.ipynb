```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix on numeric columns only
correlation_matrix = numeric_df.corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Option 1: Calculate correlation only on numerical columns
# This will automatically exclude string columns like the one with 'The Shining'
numeric_df = df.select_dtypes(include=['number'])
correlation = numeric_df.corr(method='spearman')

# Option 2: If you need to include specific string columns, you can encode them first
# For example, if you have a column of movie titles that you want to include:
from sklearn.preprocessing import LabelEncoder

# Create a copy of the dataframe to avoid modifying the original
df_encoded = df.copy()

# Identify string columns that need encoding
string_columns = df.select_dtypes(include=['object']).columns

# Apply label encoding to string columns
for col in string_columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col].astype(str))

# Now calculate correlation on the encoded dataframe
correlation = df_encoded.corr(method='spearman')
```
